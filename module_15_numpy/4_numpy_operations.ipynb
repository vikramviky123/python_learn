### Sorting, searching ...etc

```python
#numpy.sort(a, axis=-1, kind=None, order=None)
a = np.array([[1,4],
              [3,1]])
print(np.sort(a, axis=None))
print(np.sort(a, axis=1))
```

```python
#numpy.argmax(a, axis=None, out=None, *, keepdims=<no value>)
a = np.arange(6).reshape(2,3) + 10
b = np.array([x for x in range(69)])
print(a)
print(np.argmax(a, axis=0))
print(np.argmax(a, axis=1))
print(np.argmax(b, axis=0))
print(np.argmin(b, axis=0))
```

### Random number genereators

```python
# creates an array of mxn 
print("rand generates mxn matrix of random numnbers, uniform distribution over [0, 1).")
print(np.random.rand(2,5))
print("-"*80)
# creates an array of mxn 
print("randn generates mxn matrix of random numnbers, univariate “normal” (Gaussian) distribution of mean 0 and variance 1.")
print(np.random.randn(2,5))
print("-"*80)
# creates an random numbers b/w [0.0, 1.0)
print("random generates x random numbers half-open interval [0.0, 1.0) ")
print(np.random.random(5))
print("-"*80)
# creates an random integers
## random.randint(low, high=None, size=None, dtype=int) 
print("randint generates x nnumbers b/w low to high")
print(np.random.randint(2,10,30))
print("-"*80)
```

### Random distributions

* chisquare dist

 



```python
import matplotlib.pyplot as plt
#random.binomial(n, p, size=None)
n, p = 10, .5  # number of trials, probability of each trial
print("Binomial Distribution")
print(np.random.binomial(n, p, 10))
print("-"*80)
#-----------------------------------------------------------------------------------
# random.chisquare(df, size=None)
print("chisquare Distribution")
print(np.random.chisquare(2,4))
print("-"*80)
#-----------------------------------------------------------------------------------
#random.normal(loc=0.0, scale=1.0, size=None)
print("Normal Distribution")
mu, sigma = 1, 0.1 # mean and standard deviation
s = np.random.normal(mu, sigma, 10)
print(s)
print("-"*80)
#-----------------------------------------------------------------------------------
#random.poisson(lam=1.0, size=None)
import numpy as np
s = np.random.poisson(5, 10)
print(s)
```

### Gamma dist


```python
#random.gamma(shape, scale=1.0, size=None)
shape, scale = 2., 2.  # mean=4, std=2*sqrt(2)
np.random.gamma(shape, scale, 10)
```

```python
import matplotlib.pyplot as plt
import scipy.special as sps  
s=np.random.gamma(shape, scale, 1000)
print(np.mean(s))
print(np.std(s))
count, bins, ignored = plt.hist(s, 50, density=True)

y = bins**(shape-1)*(np.exp(-bins/scale) /  
                     (sps.gamma(shape)*scale**shape))
plt.plot(bins, y, linewidth=2, color='r')  
plt.show()
```
