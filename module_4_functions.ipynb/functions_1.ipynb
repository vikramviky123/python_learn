#Lambda Functions

```python
def sq(x):
    return x**2
```

```python
square_ = lambda x : x**2
```

```python
print(sq(2))
print(square_(2))
```

```python
def f4():
   global x 
   x= 200
   print("Local x:", x)
 
 
f4()
print("Global x:", x)
```

Whenever you declare a function, the non-default arguments are specified before the default arguments. The default arguments are specified at the end, meaning any arguments specified to the right of a default argument must also be a default one. It is to prevent any ambiguity by the Python interpreter. For example, the below-given code block will throw an error if we try to run it.

```python
def add_two(a=2, b):
    sum = a + b
    return sum
```

```python
def add_two(b,a=2):
    sum = a + b
    return sum
```

```python
def organizingContainers(container):
    # Write your code here
    n = len(container)
    types_sum = [0] * n
    containers_sum = [0] * n

    for i in range(n):
        containers_sum[i] = sum(container[i])
        print(f"container sum is { containers_sum[i]}")
        print("-"*25)
        for j in range(n):
            types_sum[j] += container[i][j]
            print(f"for {j}th Row: type sum is {types_sum},{container[i][j]}")

    if sorted(types_sum) == sorted(containers_sum):
        return "Possible"
    else:
        return "Impossible"
cont = [[1,3,1],[2,1,2],[3,3,3]]
cont = [[0,2,1],[1,1,1],[2,0,0]]
print(organizingContainers(cont))
```
