#Json

```python
import json

# some JSON:
x =  '{ "name":"John", "age":30, "city":"New York"}'

# parse x:
y = json.loads(x)

# the result is a Python dictionary:
print(y["age"])
```

```python
# a Python object (dict):
x = {
  "name": "John",
  "age": 30,
  "city": "New York"
}

# convert into JSON:
y = json.dumps(x)

# the result is a JSON string:
print(type(y))
print(y)
```

#date

```python
import datetime

x = datetime.datetime.now()
print(x)
print(x.year)
print(x.month)
print(x.day)
print(x.timestamp())
print(x.strftime("%A"))
print(x.strftime("%B"))
print("today's date: ", datetime.date.today())
```

```python
userdatetime = datetime.datetime(2019, 5, 10, 17, 30, 20, 154236)
userdate = datetime.date(2021, 9, 15)
usertime = datetime.time(hour=12, minute=55, second=50)

print('userdate: ', userdatetime)
print('userdate: ', userdate)
print('userdate: ', usertime)
print('type of userdate: ', type(userdate))
```

```python
#time delta

data1= datetime.timedelta(days=1, hours=23, minutes=30)
print('data1: ', data1)

data2= datetime.timedelta(days=4, hours=11, minutes=30)
print('data2: ', data2)

difference = data2-data1
print('difference between data2 and data1: ', difference)
```

#Time ZOne

```python
import datetime
import pytz


localtime= datetime.datetime.now()
print('local time: ', localtime)

utctime = datetime.datetime.now(pytz.utc)
print('UTC time zone: ', utctime)

ustime = datetime.datetime.now(pytz.timezone('US/CENTRAL'))
print('US time zone: ', ustime)
```

```python
def printTime(h,m):

    words = ['one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight', 'nine', 'ten',
    'eleven', 'twelve', 'thirteen', 'fourteen', 'fifteen', 'sixteen', 'seventeen', 'eighteen',
    'nineteen', 'twenty', 'twenty-one', 'twenty-two', 'twenty-three', 'twenty-four', 'twenty-five', 'twenty-six',
    'twenty-seven', 'twenty-eight', 'twenty-nine']

    hour_ = words[h-1]
    hour_plus = words[h]
    
    if m ==15 or m == 30 or m == 45:
        None
        
    else:       

        if m<=29:
            min_ = words[m-1]
        else:
            min_plus = words[60-(m-1)-2]


    if m == 0:
        print(hour_+" o' clock")
    if m == 60:
        print(hour_plus+" o' clock")
        
    elif m == 1:
        print(min_+" minute past "+hour_)
    
    elif m ==15:
        print("quarter past "+hour_)

    elif m ==30:
        print("half past "+hour_)
        
    elif m > 2 and m < 30:
        print(min_+" minutes past "+hour_)

    elif m ==45:
        print("quarter to "+hour_plus)
        
    elif m > 30 and m < 59:
        print(min_plus+" minutes to "+hour_plus)
        
    elif m == 59:
        print(min_plus+" minute to "+hour_plus)
```
