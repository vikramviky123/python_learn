#1. Static variables:

If the value of a variable is not varied from object to object, such type of variables we have to 
declare with in the class directly but outside of methods. Such type of variables are called Static 
variables.

For total class only one copy of static variable will be created and shared by all objects of that 
class.

We can access static variables either by class name or by object reference. But recommended to 
use class name.

```python
class Test:
    x = 20
    z = 100

    def __init__(self):
        Test.z1 = 200
        self.x = 50
        self.y = 30

    def m1(self):
        Test.z6 = 700

    @classmethod
    def m2(cls):
        Test.z2 = 300
        cls.z5 = 600

    @staticmethod
    def m3():
        Test.z3 = 400
       
```

```python
t1=Test() 
t2=Test() 
print('t1:',t1.x,t1.y,t1.z) 
print('t2:',t2.x,t2.y,t2.z) 
Test.x=888 
t1.y=999 
print('t1:',t1.x,t1.y) 
print('t2:',t2.x,t2.y) 
t1.m1()
t1.m2()
Test.z4 = 500
print("Printing statuc variables")
print(Test.z1)
print(Test.z2)
print(Test.z3)
print(Test.z4)
print(Test.z5)
print("--------------------")
print(t1.z4)

print(Test.__dict__)
```

How to access static variables:

1. inside constructor: by using either self or classname
2. inside instance method: by using either self or classname
3. inside class method: by using either cls variable or classname
4. inside static method: by using classname
5. From outside of class: by using either object reference or classnmae


```python
import re
import sys

class newVikramBank:
    """
    This is my New bank I am opening in my colony,
    Please wish me a great luck and success going forward.

    Ill take envelopes if you got any-->:)

    THNAK YOU FOR VISITING
    
    """
    bankName = "New Vikram Bank founded 1857"

    def __init__(self, name, balance=0.0):
        self.name    = name 
        self.balance =balance

    def deposit(self, amt):
        self.balance = self.balance + amt
        print("-"*50)
        print("Balance after deposit :",self.balance)
        print("-"*50)
        print()


    def withdraw(self, amt):
        if amt > self.balance:
            sys.exit()
            print("Insufficient Funds --> PLEASE COME BACK LATER")
        else:               
            self.balance = self.balance - amt
            print("-"*50)
            print("Balance after withdrawal :",self.balance)
            print("-"*50)
            print()

print("welcome to :",newVikramBank.bankName)
print("Choose what you came for")

name = input("Enter your name here:")
c1 = newVikramBank(name)
print('\n')

while True:
    
    print("Please Choose one of the options below")
    print('d-Deposit\nw-Withdraw\ne-Exit')

    option = input("Choose your Option :")
    if re.findall("[dewDEW]",option):
        option = option.lower()

        if option == 'd':
            amt = float(input("Enter amount to be Deposited:"))
            c1.deposit(amt)

        elif option == 'w':
            amt = float(input("ENter amount to be Withdrawn:"))
            c1.withdraw(amt)


        elif option == 'e':
            print("-"*50)
            print("You have chosen to exit--->","Thank you for visiting".upper())
            print("                            SEE YA")
            print("-"*50)
            break
    else:
        print("-"*80)
        print("Enter Valid Options in small case d or e or w")
        print("-"*50)
        continue
```
