```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df_poll = pd.read_csv("Delhi_Pollution.csv")
df_uber = pd.read_csv("train-data.csv")
df_uber.pop("Unnamed: 0")
df_poll.head(2)
df_uber.head(2)
```

```python
font_title = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 24,
        }
font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 16,
        }
fig, ax = plt.subplots(figsize =(20,5), layout='constrained')
ax.plot(df_poll["PM2.5"], label ="PM2.5", 
        color = "green", marker = "o", 
        linestyle = "--", linewidth = 5, markersize = 0.5)# color, marker, linestyle, linewidth,markersize
ax.plot(df_poll["PM10"], label ="PM10" )


ax.set_title("Particulate Matter PM2.5 and PM10", fontdict=font_title)
ax.set_xlabel("Particulate Matter",fontdict=font)
ax.set_ylabel("PPM",fontdict=font)
ax.text(10,600,"Just Fun", fontdict=font_title)
ax.tick_params(labelcolor = "brown", labelsize = 12, width = 4,)
ax.legend()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

font_title = {
    'family': 'serif',
    'color':  'darkred',
    'weight': 'normal',
    'size': 24,
}
font = {
    'family': 'serif',
    'color':  'darkred',
    'weight': 'normal',
    'size': 16,
}

# Set Seaborn style and context
sns.set_style("whitegrid")
sns.set_context("notebook", font_scale=1.2, rc={"lines.linewidth": 2.5})

fig, ax = plt.subplots(figsize=(20, 5))

# Plotting using seaborn
sns.lineplot(data=df_poll, x=df_poll.index, y="PM2.5", ax=ax, label="PM2.5", color="green", marker="o", linestyle="--", linewidth=5, markersize=0.5)
sns.lineplot(data=df_poll, x=df_poll.index, y="PM10", ax=ax, label="PM10", linewidth=1,color="darkblue")

ax.set_title("Particulate Matter PM2.5 and PM10", fontdict=font_title,color='0.2')
ax.set_xlabel("Particulate Matter", fontdict=font)
ax.set_ylabel("PPM", fontdict=font)
ax.text(10, 600, "Just Fun", fontdict=font_title)
ax.tick_params(labelcolor="blue", labelsize=12, width=4)

ax.legend()

plt.show()
```
